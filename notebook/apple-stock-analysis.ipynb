````python
# Or, we can plot the four essential plots all at once:

def tsplot(y, lags=None, title='', figsize=(14, 8)):
    '''Examine the patterns of ACF and PACF, along with the time series plot and histogram.
    
    Original source: https://tomaugspurger.github.io/modern-7-timeseries.html
    '''
    fig = plt.figure(figsize=figsize)
    layout = (2, 2)
    ts_ax   = plt.subplot2grid(layout, (0, 0))
    hist_ax = plt.subplot2grid(layout, (0, 1))
    acf_ax  = plt.subplot2grid(layout, (1, 0))
    pacf_ax = plt.subplot2grid(layout, (1, 1))
    
    y.plot(ax=ts_ax)
    ts_ax.set_title(title)
    y.plot(ax=hist_ax, kind='hist', bins=25)
    hist_ax.set_title('Histogram')
    smt.graphics.plot_acf(y, lags=lags, ax=acf_ax)
    smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax)
    [ax.set_xlim(0) for ax in [acf_ax, pacf_ax]]
    sns.despine()
    plt.tight_layout()
    return ts_ax, acf_ax, pacf_ax
````

````python
print("Apple's stock price:", "\n", AAPL.dtypes, "\n")
````

````python
# explore the first 10 observations
AAPL.head(10)
````

````python
AAPL_short = AAPL.Close['2005-01-01':]
````

````python
# Visualise AAPL's stock price
yr_moving_avg = AAPL_short.rolling(365).mean()
mnth_moving_avg = AAPL_short.rolling(30).mean()


AAPL_short.plot(figsize=(12,8))
yr_moving_avg.plot(figsize=(12,8), color='red', label="Yearly avg")
mnth_moving_avg.plot(figsize=(12,8), color='turquoise', label="Monthly avg")
#plt.plot(moving_avg, color='red')
plt.legend(bbox_to_anchor=(1.25, 0.5))
plt.title("APPLE stock price from 2005 to current")
sns.despine()
````

````python
diff_AAPL_short = AAPL_short.diff(periods=1)
````

````python
diff_AAPL_short.plot(figsize=(12,9))
plt.legend(bbox_to_anchor=(1.25, 0.5))
sns.despine()
````

````python
ln_AAPL_short = np.log(AAPL_short)
````

````python
ln_AAPL_short.plot(figsize=(12,9))
plt.legend(bbox_to_anchor=(1.25, 0.5))
sns.despine()
````

````python
fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ln_AAPL_short, lags=100, ax=ax1)
fig.tight_layout();

ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ln_AAPL_short, lags=100, ax=ax2)
ax2.xaxis.set_ticks_position('bottom')
fig.tight_layout();
````

````python
Monthy_AAPL_short = pd.DataFrame(AAPL_short.groupby(by=AAPL_short.index.strftime('%y-%b')).mean())
````

````python
Monthy_AAPL_short.index = pd.to_datetime(Monthy_AAPL_short.index,
                                         yearfirst=True,
                                        format="%y-%b")
````

````python
Monthy_AAPL_short.plot(figsize=(12,9))
plt.legend(bbox_to_anchor=(1.25, 0.5))
plt.title("Monthly Average price Apple Stock")
sns.despine()
fig.tight_layout();
````

````python
ln_Monthy_AAPL_short = np.log(Monthy_AAPL_short)


ln_Monthy_AAPL_short.plot(figsize=(12,9))
plt.legend(bbox_to_anchor=(1.25, 0.5))
plt.title("Monthly Average price Apple Stock")
sns.despine()
fig.tight_layout();
````

````python
fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ln_Monthy_AAPL_short, lags=20, ax=ax1)
fig.tight_layout();

ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ln_Monthy_AAPL_short, lags=20, ax=ax2)
ax2.xaxis.set_ticks_position('bottom')
fig.tight_layout();
````

````python
# Scatterplot matrix is another way to visualize the autocorrelation
# Its advantage is that it is very intuitive, as scatterplot (i.e. one of the plots in a scatterplot matrix) 
# is used often in practice

lags=9

ncols=3
nrows=int(np.ceil(lags/ncols))

fig, axes = plt.subplots(ncols=ncols, nrows=nrows, figsize=(4*ncols, 4*nrows))

for ax, lag in zip(axes.flat, np.arange(1,lags+1, 1)):
    lag_str = 't-{}'.format(lag)
    X = (pd.concat([ln_Monthy_AAPL_short, ln_Monthy_AAPL_short.shift(-lag)], axis=1,
                   keys=['y'] + [lag_str]).dropna())

    X.plot(ax=ax, kind='scatter', y='y', x=lag_str);
    corr = X.corr().as_matrix()[0][1]
    ax.set_ylabel('Original')
    ax.set_title('Lag: {} (corr={:.2f})'.format(lag_str, corr));
    ax.set_aspect('equal');
    sns.despine();

fig.tight_layout();
````
